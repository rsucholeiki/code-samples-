## Part 1: numeric solution of the linear population model

Consider the linear difference equation below, where $N$ is the population vector and $r$ is the growth rate, expressed by the following iterative discrete-time model:

$$N(t+1)=rN(t)$$

This can represent a population that changes by an equal proportion every time step, e.g. if it grows by 5% a year, then $r=1.05$ and $t$ is time in years.

**Q1.1:** Let the population $N$ start with 700 individuals, with a constant mulitplicative growth rate of $r=0.8$. Use a for loop to propagate this dynamical systems for 20 timesteps, saving all the values into the array variable $N$. Then create another variable $t$ to represent time and plot $N$ as a function of time using the plt.plot commmand.

Self-check: the population after 20 iterations should be approximately 8.

```python
numsteps = 20 #number of time steps
rate = .8 #define population change rate
N = np.zeros(numsteps + 1) #Create a solution array for population values
N[0] = 700 #Set initial population as 700 individuals
for i in range(numsteps): #Create a for loop to propagate the system for 20 timesteps
    N[i+1] = rate*N[i]
t = np.arange(numsteps + 1) #Create a time vector
plt.plot(t,N) #Plot N as a function of time
plt.xlabel('time')
plt.ylabel('population')
title = 'Population as a function of time for multiplicative growth rate r = ' + str(rate)
plt.title(title)
plt.show()
```

The value of r determines whether the population undergoes growth or decay at increasing values of time. When r is less than 1, the population decays and tends closer to zero at large values of time. When r is greater than 1, the population increases with increasing time. If r = 1, then the population remains constant throughout time. Furthermore, the magnitude by which r differs from 1 determines the rate of change; for example, the linear population model with r = .8 from the previous part decays more quickly than r = .98. 

## Part 2: linear model with a constant additive term

Consider the modified linear growth model with a constant additive term $a$, and the other variables and parameters with the same meaning:

$$ N(t+1)=rN(t)+a $$

This may represent a population with a constant migration or death rate (e.g. in a controlled population like an aquarium or a nature preserve where a set number of hunting permits is issued.)

**Q2.1:** Let the population $N$ have the growth rate of $r = 0.95$ and the constant rate $a = 10$.
```python
#N(0) = 0
numsteps = 100 #number of time steps
rate = .95 #define population change rate
a = 10 #define the constant rate
N = np.zeros(numsteps + 1) #Create a solution array for population values
N[0] = 0 #Set initial population as 50 individuals
for i in range(numsteps): #Create a for loop to propagate the system for 100 `timesteps
    N[i+1] = rate*N[i] + a
t = np.arange(numsteps + 1) #Create a time vector
plt.plot(t,N, label = 'N(0) = '+str(N[0])) #Plot N as a function of time
plt.xlabel('time')
plt.ylabel('population')
title = 'Solutions for linear models with constant additive term'
plt.title(title)

#N(0) = 400
numsteps = 100 #number of time steps
rate = .95 #define population change rate
a = 10 #define the constant rate
N = np.zeros(numsteps + 1) #Create a solution array for population values
N[0] = 400 #Set initial population as 400 individuals
for i in range(numsteps): #Create a for loop to propagate the system for 100 timesteps
    N[i+1] = rate*N[i] + a
t = np.arange(numsteps + 1) #Create a time vector
plt.plot(t,N, label = 'N(0) = '+str(N[0])) #Plot N as a function of time

#N(0) = 800
numsteps = 100 #number of time steps
rate = .95 #define population change rate
a = 10 #define the constant rate
N = np.zeros(numsteps + 1) #Create a solution array for population values
N[0] = 800 #Set initial population as 800 individuals
for i in range(numsteps): #Create a for loop to propagate the system for 100 timesteps
    N[i+1] = rate*N[i] + a
t = np.arange(numsteps + 1) #Create a time vector
plt.plot(t,N, label = 'N(0) = '+str(N[0])) #Plot N as a function of time

plt.legend()
plt.show()
```

When we vary the starting population size, we observe that all populations end up converging to a common size at around 200 individuals at large values of time. This is due to the constant additive term in our linear growth models. The phenomenon is due to the fact that when N(t) = 200, N(t+1) = rN(t) + a = .95(200) + 10 = 200. Therefore, for these linear models, a population size of around 200 individals represents a stable fixed point that is reached no matter the starting size.

## Part 3: linear model with a random growth rate

Let us consider a more realistic model of a population with a linear growth rate than the deterministic (predictable) model with constant growth rate $r$: $N(t+1) = r N(t)$. In reality, no population has the exact same growth rate every cycle (whether it's a year or an hour) - there is always some unpredictable variation, called stochasticity. Let us assume that the population has the mean growth rate of 1.02 (or 2%) per year but this value fluctuates with a normal distribution, which means it's equally probable to be higher or lower than the mean. The big question is:

**Will the population with a fluctuating growth rate with mean value r grow faster, slower, or the same as the population with the constant growth rate r?**

Q3.1: Use your population model script from part 1 to simulate this situation, adding a *normal random number generator* with the mean of 1.02 a standard deviation of your choice to generate the growth rate r and iterate the population for 500 steps. Important hint: since we want to keep the growth rate positive, keep the standard deviation small enough that the random number is always positive. Generate the solution and plot its graph, and plot the solution for the deterministic population in the same figure. Can you make a guess about the answer of the question from this plot?

```python
#Deterministic linear model (Part I)
numsteps = 500 #number of time steps
rate = 1.02 #define population change rate
N = np.zeros(numsteps + 1) #Create a solution array for population values
N[0] = 700 #Set initial population as 700 individuals
for i in range(numsteps): #Create a for loop to propagate the system for 20 timesteps
    N[i+1] = rate*N[i]
t = np.arange(numsteps + 1) #Create a time vector
plt.plot(t,N, label = 'Deterministic model') #Plot N as a function of time
plt.xlabel('time')
plt.ylabel('population')
title = 'Population as a function of time for stochastic and deterministic models'
plt.title(title)

#Stochastic linear model
numsteps = 500 #number of time steps
#Define parameters for the stochastic growth rate
mu = 1.02
sigma = 0.05
num = numsteps 
rate = np.random.normal(mu, sigma, num)

N = np.zeros(numsteps + 1) #Create a solution array for population values
N[0] = 700 #Set initial population as 700 individuals

for i in range(numsteps): #Create a for loop to propagate the system for 500 timesteps
    N[i+1] = rate[i]*N[i]
t = np.arange(numsteps + 1) #Create a time vector
plt.plot(t,N, label = 'Stochastic model') #Plot N as a function of time    

plt.legend()
plt.show()
```

After plotting the linear growth models with both stochastic and deterministic growth rates and running the simulation several times, it appears that the stochastic model sometimes grow faster and sometimes grows slower than the deterministic model. Of course, the stochastic model ending population never exactly matches the deterministic model. But because due to the normal distribution of growth rates generated around a mean of 1.02, the stochastic model does not exclusively grow faster or slower than the deterministic model but may vary either way on different runs. 

As an example, when sigma = .05 for the stochastic rate generator, I ran the stochastic model 10 times. The results are as follows: the stochastic model grew FASTER than the deterministic model 2 times, and SLOWER than the deterministic model 8 times. From this limited sample, it appears possible that the stochastic model grows slower, on average, than the deterministic model.

However, from only running one simulation for this limited number of times, it is still not completely clear whether or not growth is faster or slower on average. Q3.2: Since the results of the simulation will be different each time you run it, let us repeat it multiple times and look at the mean behavior of all the simulated populations.
```python
#Deterministic linear model (Part I)
numsteps = 500 #number of time steps
rate = 1.02 #define population change rate
N = np.zeros(numsteps + 1) #Create a solution array for population values
N[0] = 700 #Set initial population as 700 individuals
for i in range(numsteps): #Create a for loop to propagate the system for 20 timesteps
    N[i+1] = rate*N[i]
t = np.arange(numsteps + 1) #Create a time vector
plt.plot(t,N) #Plot N as a function of time
plt.xlabel('time')
plt.ylabel('population')
title = 'Population as a function of time for stochastic and deterministic models'
plt.title(title)

#Print the population size after 500 time steps for the deterministic model
string1 = 'The population size of the deterministic model after 500 time steps is ' + str(N[500])
print(string1)

#Stochastic linear model
numsimulations = 100 #Set number of simulations
numsteps = 500 #number of time steps
#Define parameters for the stochastic growth rate
mu = 1.02
sigma = 0.05

poparray = np.zeros(numsimulations) #Make an array to save the population values after 500 time steps

for j in range(numsimulations): #Create a for loop to create plots
    rate = np.random.normal(mu, sigma, numsteps)
    N = np.zeros(numsteps + 1) #Create a solution array for population values
    N[0] = 700 #Set initial population as 700 individuals
    for i in range(numsteps): #Create a for loop to propagate the system for 500 timesteps
        N[i+1] = rate[i]*N[i]   
    plt.plot(t,N) #Plot N as a function of time
    poparray[j] = N[numsteps]

string2 = 'The mean population size of the stochastic model after 500 time steps (average over ' + str(numsimulations) + ' simulations) is ' + str(np.mean(poparray))
print(string2)

string3 = 'The standard deviation of population size for the stochastic model after 500 time steps is ' + str(np.std(poparray))
print(string3)

string4 = 'The median population size for the stochastic model after 500 time steps (average over ' + str(numsimulations) + ' simulations) is ' + str(np.median(poparray))
print(string4)

plt.show()
```
